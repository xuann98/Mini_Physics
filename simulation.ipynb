## Newton's Gravitational Law
$ F = G\cdot \frac{m_1m_2}{r^2}$

We assume that $m_1$ is the mass of the large 'fixed' object (i.e the sun).

Newton's second law is given by $F_1 = m_1a$

By subbing in the newton's second law into the gravitational law, we can find the acceleration of the other body by
$a_j = G \cdot \frac{m_j}{r_{ij}^2}$

Velocity and distance are computed with the simple Euler method, a first-order numerical procedure for solve ordinary differential equations with a given initial value.

```python
def compute_forces(bodies):
    """
    Computing gravitational acceleration on each body due to the other bodies

    """
    for body in bodies:
        body.reset_acceleration() # resets acceleration after each step
    
    # compute pairwise acclerations
    for i, body_i in enumerate(bodies):
        for j, body_j in enumerate(bodies):
            if i == j:
                continue # no gravitational pull on itself
            
            # vector from i to j
            r_vec = body_j.position - body_i.position
            r_mag = np.linalg.norm(r_vec) # normalizing the position vector and gives only its magnitude
            if r_mag == 0:
                continue # no division by zero
            
            #gravitational acceleration magnitude
            a_mag = G * body_j.mass / r_mag**2
            
            #acceleration vector
            body_i.acceleration += a_mag * (r_vec / r_mag) # the expression in the bracket is the direction unit vector 
```

```python
def simulate(bodies, dt, steps):
    """"
    Run the simulation for a given number of time steps
    
    """
    for step in range(steps):
        compute_forces(bodies) #update acceleration
        for body in bodies:
            body.update_velocity(dt)
            body.update_position(dt)
```

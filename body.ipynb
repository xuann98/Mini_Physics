Defining what a planet/start object is.

Each body constains
- mass
- position (x,y)
- velocity 
- acceleration


```python
import numpy as np
```

```python
class Body: #just to store state of the object
    def __init__(self, name, mass, position, velocity, color = "white", size = 5):
        self.name = name
        self.mass = mass
        self.position = np.array(position, dtype= float)
        self.velocity = np.array(velocity, dtype=float)
        # acceleration computed during simulation so here we just initialise it
        self.acceleration = np.zeros(2)
        #for visualisation purposes
        self.color = color
        self.size = size
        # to store previous positions for orbit trails
        self.trail = []
    
    def reset_acceleration(self):
        """
            Reset acceleration before reculating the forces
            
        """
        self.acceleration = np.zeros(2)

    def update_velocity(self, dt):
        """
        Euler update: velocity = velocity + acceleration * dt
        
        """
        self.velocity += self.acceleration * dt

    def update_position(self, dt):
        """
        Euler update: position = position + velocity * dt, store trail
        
        """
        self.position += self.velocity * dt
        self.trail.append(self.position.copy())
   


```
